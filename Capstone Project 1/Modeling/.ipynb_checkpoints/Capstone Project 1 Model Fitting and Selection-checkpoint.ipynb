```python
#Import libraries
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

from statsmodels.graphics.gofplots import ProbPlot
```

```python
#Set style
plt.style.use('seaborn') # pretty matplotlib plots
```

```python
#Import packages for hyper parameter tuning 

from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import explained_variance_score
from sklearn.model_selection import KFold
from xgboost import XGBRegressor
import xgboost 
```

```python
#Read in data - this analysis will use log-transformed outcomes

df = pd.read_csv('C:/Users/robva/Dropbox (Personal)/Springboard Data Science Course/Capstone Project 1/EDA/file_for_analysis_1.csv')   
```

```python
#Transform the outcomes
df['visitors_log1p'] = np.log1p(df['visitors'])
df['visitors_recode_log1p'] = np.log1p(df['visitors_recode'])
```

```python
#Create log-transformed rolling means, medians, and standard deviations by restaurant for each 30 day window.  
#Allow calculations for cells where less than 30 observations are available by setting min_periods = 1.
df['rolling_mean_log1p'] = df.groupby('air_store_id')['visitors_recode_log1p'].rolling(30, min_periods=1).mean().reset_index(0,drop=True)
df['rolling_median_log1p'] = df.groupby('air_store_id')['visitors_recode_log1p'].rolling(30, min_periods=1).median().reset_index(0,drop=True)
df['rolling_std_log1p'] = df.groupby('air_store_id')['visitors_recode_log1p'].rolling(30, min_periods=1).std().reset_index(0,drop=True)
#Std isn't being calculated for about 800 observations so I'm going to fill NaN's with the overall std for each restaurant
df['rolling_std_log1p'] = df.groupby(['air_store_id'])['rolling_std_log1p'].fillna(df['visitors_recode_log1p'].std()).reset_index(0,drop=True)
```

```python
#Create one week lags for log-transformed rolling means, medians, and standard deviations and replace NaN's with the mean of each by restaurant
df['rolling_mean_log1p_lag'] = df.groupby('air_store_id')['rolling_mean_log1p'].shift(7).fillna(df['visitors_recode_log1p'].mean()).reset_index(0,drop=True)
df['rolling_median_log1p_lag'] = df.groupby('air_store_id')['rolling_median_log1p'].shift(7).fillna(df['visitors_recode_log1p'].median()).reset_index(0,drop=True)
df['rolling_std_log1p__lag'] = df.groupby('air_store_id')['rolling_std_log1p'].shift(7).fillna(df['visitors_recode_log1p'].std()).reset_index(0,drop=True)
```

```python
#Create one week lag for each restaurant's log-transformed visitor count and replace missing with mean number of visitors for that restaurant.
df['visitors_log1p_lag'] = df.groupby('air_store_id')['visitors_recode_log1p'].shift(7).fillna(df['visitors_recode_log1p'].mean()).reset_index(0,drop=True)
```

```python
#Create training and test sets

df['visit_date'] = pd.to_datetime(df['visit_date'])

#fix this to make it inclusive - the problem is in the string - look into between datetime function

train = df.loc[df['visit_date'] <= '2017-3-27']
test = df.loc[df['visit_date'] > '2017-3-27']

```

```python
#Create features by restaurant on the training set to avoid leakage from test set

mean_visitors = train.groupby(['air_store_id'])['visitors'].mean().reset_index()
mean_visitors.columns = ['air_store_id','mean_visitors']
train = train.merge(mean_visitors, how = 'left', on = 'air_store_id')
test = test.merge(mean_visitors, how = 'left', on = 'air_store_id')
```

```python
#Get descriptives for mean_visitors
train['mean_visitors'].describe()
```

```python
#Create dummies for low-,mid-, and high-traffic restaurants
train['low_traffic'] = (train['mean_visitors'] <=9.92).astype(int)
train['mid_traffic'] = ((train['mean_visitors'] > 9.75) & (train['mean_visitors'] < 24.83)).astype(int)
train['high_traffic'] = (train['mean_visitors'] >= 24.83).astype(int)

test['low_traffic'] = (test['mean_visitors'] <=9.92).astype(int)
test['mid_traffic'] = ((test['mean_visitors'] > 9.75) & (train['mean_visitors'] < 24.83)).astype(int)
test['high_traffic'] = (test['mean_visitors'] >= 24.83).astype(int)
```

```python
#Create mean visitors by day by restaurant - perhaps this will capture the mispredictions on days with zero visitors (days closed?)

mean_visitors_by_day = train.groupby(['air_store_id', 'day_of_week'])['visitors'].mean().reset_index()
mean_visitors_by_day.columns = ['air_store_id', 'day_of_week','mean_visitors_by_day']
train = train.merge(mean_visitors_by_day, how = 'left', on = ['air_store_id','day_of_week'])
test = test.merge(mean_visitors_by_day, how = 'left', on = ['air_store_id','day_of_week'])
```

```python
#Create log-transformed mean visitors by restaurant 

mean_visitors_log1p = train.groupby(['air_store_id'])['visitors_log1p'].mean().reset_index()
mean_visitors_log1p.columns = ['air_store_id','mean_visitors_log1p']
train = train.merge(mean_visitors_log1p, how = 'left', on = 'air_store_id')
test = test.merge(mean_visitors_log1p, how = 'left', on = 'air_store_id')
```

```python
#Get descriptives for log-transformed mean_visitors
train['mean_visitors_log1p'].describe()
```

```python
#Create dummies for low-,mid-, and high-traffic restaurants using log-transformed outcome
train['low_traffic_log1p'] = (train['mean_visitors_log1p'] <=1.97).astype(int)
train['mid_traffic_log1p'] = ((train['mean_visitors_log1p'] > 1.97) & (train['mean_visitors_log1p'] < 2.86)).astype(int)
train['high_traffic_log1p'] = (train['mean_visitors_log1p'] >= 2.86).astype(int)

test['low_traffic_log1p'] = (test['mean_visitors_log1p'] <=1.97).astype(int)
test['mid_traffic_log1p'] = ((test['mean_visitors_log1p'] > 1.97) & (test['mean_visitors_log1p'] < 2.86)).astype(int)
test['high_traffic_log1p'] = (test['mean_visitors_log1p'] >= 2.86).astype(int)
```

```python
#Create log-transformed mean visitors by day by restaurant 
mean_visitors_by_day_log1p = train.groupby(['air_store_id', 'day_of_week'])['visitors_log1p'].mean().reset_index()
mean_visitors_by_day_log1p.columns = ['air_store_id', 'day_of_week','mean_visitors_by_day_log1p']
train = train.merge(mean_visitors_by_day_log1p, how = 'left', on = ['air_store_id','day_of_week'])
test = test.merge(mean_visitors_by_day_log1p, how = 'left', on = ['air_store_id','day_of_week'])
```

```python
#To make the test set more like real-world data, I will make the target variable the original data with outliers included.
#I will still train on data with outliers replaced with capped values

train = train.drop('visitors_log1p', axis=1)
test = test.drop('visitors_recode_log1p', axis=1)
train = train.rename(columns={'visitors_recode_log1p':'visitors_log1p'})
```

```python
#Drop variables that won't be used in the analysis
train = train.drop(['Unnamed: 0', 'latitude','longitude','air_store_id', 'air_genre_name','air_area_name','day_of_week','station_id',
              'avg_temperature_missing', 'high_temperature_missing', 'low_temperature_missing', 'precipitation_missing',
             'hours_sunlight_missing', 'extreme_value', 'rolling_mean', 'rolling_median', 'rolling_std', 'visitors', 'visitors_recode', 
              'rolling_mean_log1p','rolling_median_log1p','rolling_std_log1p'], axis=1)

test = test.drop(['Unnamed: 0', 'latitude','longitude','air_store_id', 'air_genre_name','air_area_name','day_of_week','station_id',
              'avg_temperature_missing', 'high_temperature_missing', 'low_temperature_missing', 'precipitation_missing',
             'hours_sunlight_missing', 'extreme_value', 'rolling_mean', 'rolling_median', 'rolling_std', 'visitors', 'visitors_recode', 
              'rolling_mean_log1p','rolling_median_log1p','rolling_std_log1p'], axis=1)
```

```python
#Create visit date keys and drop from training and test data

train_dates = train['visit_date']
test_dates = test['visit_date']
train = train.drop('visit_date',axis=1)
test = test.drop('visit_date',axis=1)
```

```python
#Create training and test features and targets

Xtrain = train.drop('visitors_log1p', axis=1)
Xtest = test.drop('visitors_log1p', axis=1)
ytrain = train['visitors_log1p']
ytest = test['visitors_log1p']

```

```python
#Set up hyperparameter grid
params = {
        'learning_rate': [.2, .5, .7],
        'max_depth': [5,7,9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [.2,.4,0.6],
        'n_estimators': [20,30,40],
        'alpha': [5,10,15]
        }
```

```python
#Instantiate the regressor
xgb = XGBRegressor(objective='reg:linear')
```

```python
 #Use random grid search to tune parameters
clf = RandomizedSearchCV(estimator=xgb, param_distributions=params, n_iter=10, random_state=42)
clf.fit(Xtrain,ytrain)
print(clf.best_params_)
print(clf.best_score_)
```

```python
#Reset hyperparameter grid to reexamine values on the edge of the grid
params = {
        'max_depth': [2,3,4,5,7,9]
        }
```

```python
#Reinstantiate regressor
xgb = XGBRegressor(objective='reg:linear',subsample = 0.8, n_estimators = 30,learning_rate = 0.5, colsample_bytree = 0.4, alpha = 10, random_state=42)
```

```python
#Use random grid search to tune parameters
clf = RandomizedSearchCV(estimator=xgb, param_distributions=params, n_iter=6, random_state=42)
clf.fit(Xtrain,ytrain)
print(clf.best_params_)
print(clf.best_score_)
```

```python
#Reinstantiate regressor
xgb = XGBRegressor(objective='reg:linear',subsample = 0.8, max_depth = 7, n_estimators = 30,learning_rate = 0.5, colsample_bytree = 0.4, alpha = 10, random_state = 42)
```

```python
#Train the model
xgb.fit(Xtrain,ytrain)
```

```python
#Get predictions
preds = xgb.predict(Xtest)
```

```python
#Get error using rmse
rmse = np.sqrt(mean_squared_error(ytest, preds))
print("RMSE: %f" % (rmse))
```

```python
#Compare with training RMSE to see if there is overfitting
preds_train = xgb.predict(Xtrain)
#Get error using rmse
rmse = np.sqrt(mean_squared_error(ytrain, preds_train))
print("RMSE: %f" % (rmse))
```

```python
#Create fitted value dataset

ytest = pd.DataFrame(ytest)
preds = pd.DataFrame(preds)
df_target = pd.DataFrame(np.concatenate((ytest, preds), axis=1,))
df_target.columns = ['ytest_log1p','preds'] 
df_target['resids'] = df_target['ytest_log1p'] - df_target['preds']
```

```python
#Examine and calculate residuals 

plt.clf()
plt.rcParams['figure.figsize'] = [5, 5]
sns.regplot(df_target['preds'], df_target['resids'], lowess=True, line_kws={'color': 'red'})
plt.xlabel('Fitted Values')
plt.ylabel('Unstandardized Residuals')
plt.show()
```

```python
#Construct a Q-Q plot 

from sklearn import preprocessing
plt.clf()
plt.rcParams['figure.figsize'] = [5, 5]
df_target['std_resids'] = preprocessing.scale(df_target['resids'])
QQ = ProbPlot(df_target['std_resids'])
plot_lm_2 = QQ.qqplot(line='45', alpha=0.5, color='#4C72B0', lw=1)
plt.show()
```

```python
#Examine feature importances
plt.clf()
plt.rcParams['figure.figsize'] = [20, 20]
xgboost.plot_importance(xgb)
plt.show()
```

```python
# Get non-log-transformed RMSE
df_target['raw_test'] = np.expm1(df_target['ytest_log1p'])
df_target['raw_preds'] = np.expm1(df_target['preds'])
#Get error using rmse
rmse = np.sqrt(mean_squared_error(df_target['raw_test'], df_target['raw_preds']))
print("RMSE: %f" % (rmse))
```

```python
#Calculate median-estimated RMSE
df_target['median'] = df_target['raw_test'].median()
rmse = np.sqrt(mean_squared_error(df_target['raw_test'], df_target['median']))
print("RMSE: %f" % (rmse))
```

```python
#There are still a handful of extreme outliers.  
#To get a better sense of model performance on unseen data, I'm going to remove the two outliers with values greater than 200 and reassess fit.

#Create dataframe that includes both fitted and observed values as well as residuals

df_target['resids'] = df_target['raw_test'] - df_target['raw_preds']

#Drop two extreme outliers
df_target = df_target.loc[df_target['resids'] < 200]
```

```python
#Get error using rmse
rmse = np.sqrt(mean_squared_error(df_target['raw_test'], df_target['raw_preds']))
print("RMSE: %f" % (rmse))
```

```python
#Calculate median-estimated RMSE
rmse = np.sqrt(mean_squared_error(df_target['raw_test'], df_target['median']))
print("RMSE: %f" % (rmse))
```
